#### Import Libraries

```python
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
import time
```

```python
def get_content(link):
  '''
  A function created to scrape news article from Online Khabar Website and save
  extracted text in a file.

  Input:
  link: Homepage link to Online Khabar.
  '''

  nav_links = []
  i = 1

  # first get the links of sub topics from navigation bar: politics, economy..
  header = requests.get(link).text
  soup = BeautifulSoup(header, 'lxml')
  nav = soup.find('nav')


  nav_bar = nav.find_all('a')
  for link in nav_bar:
    nav_links.append(link.get('href'))

  # scrape the article links through sub topics page
  for link in tqdm(nav_links):
    print('\nGetting Sub Page Links...')
    sub_page = requests.get(link).text
    soup = BeautifulSoup(sub_page, 'lxml')

    news_post = soup.find_all('div', class_=['ok-news-post', 'ltr-post'])

    # scrape the actual article post
    article_links = []
    for post in news_post:
      article_links.append(post.a.get('href'))

    text = ''
    num = 1
    for article_link in article_links:
      print(f'\nGetting Article {num}...')
      article_content = requests.get(article_link).text
      soup = BeautifulSoup(article_content, 'lxml')

      article_container = soup.find('div', class_='post-content-wrap')
      article_p = article_container.find_all('p')
      for p in article_p[1:]:
        text += p.text

      num += 1
      time.sleep(8)

    print('\nSaving File...')
    # store the extracted text in a file
    with open(f'Data/sub_topic_{i}.txt', 'w') as f:
      f.write(text)

    i += 1

    print('\nDone...')
```

```python
LINK = 'https://english.onlinekhabar.com/'
get_content(LINK)
```
